```python
# Import dependencies 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import os 
```

```python
# Load CSV into a Data Frame
data = pd.read_csv("Shopify_Challenge.csv")
df = pd.DataFrame(data)
```

```python
# Get summary statistics of "Order Amount" Column
df.order_amount.describe()
```

```python
# Plot the "Order Amount" Column
df.boxplot(column='order_amount')
plt.title('Order Amount Boxplot')
plt.tight_layout()
```

```python
# Zoom in 
df.boxplot(column='order_amount')
plt.title('Order Amount Boxplot')
plt.ylim(0, 1000)
```

```python
# Investigate outliers 
unique_values = df.groupby(['order_amount']).size().reset_index(name='count').sort_values(by = 'order_amount', ascending=False)
unique_values.head(15)
```

```python
# Show high frequency amount rows 
pd.options.display.max_rows
frequency = df.loc[df['order_amount'].isin([704000, 77175, 51450, 25725])].sort_values(by='order_amount', ascending=False)
frequency
```

```python
# Use 1.5 IQR rule to truncate dataset

q1 = df.order_amount.quantile(q=0.25)
q2 = df.order_amount.quantile(q=0.5)
q3 = df.order_amount.quantile(q=0.75)
IQR = q3 - q1 

truncated_df = df[(df.order_amount < q3 + 1.5 * IQR) & (df.order_amount > q1 - 1.5 * IQR)]
truncated_df.order_amount.describe()
```

```python
# Plot the truncated Data Frame
truncated_df.boxplot(column='order_amount')
plt.title('Truncated Order Amount Boxplot')
plt.tight_layout()
plt.ylim(0, 731)
```
